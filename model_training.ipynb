```python
from sklearn.preprocessing import LabelEncoder  #to convert into categeorical data to numerical data
le = LabelEncoder()
df['churn'] = le.fit_transform(df['churn'])
```

```python
df
```

```python
df.info()
```

```python
cat_cols=df.select_dtypes('object') #the data of obj we convert into numerical
```

```python
from sklearn.preprocessing import LabelEncoder 
encoder=LabelEncoder()

for col in cat_cols:
    df[col]=encoder.fit_transform(df[col])  #converted data here to numerical
```

```python
df
```

```python
df.drop(columns=['customer_id'], inplace=True)
```

```python
df
```

```python
x=df.drop(columns=['churn'])
y=df['churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(x_train, y_train)
```

```python
y_pred=model_lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt= DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
y_pred=model_dt.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC

model_svm = SVC()
model_svm.fit(x_train, y_train)
```

```python
y_pred=model_svm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```python
y_pred=model_knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
